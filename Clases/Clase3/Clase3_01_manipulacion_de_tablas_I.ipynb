La función filter() de dplyr es el equivalente a la sentencia where de sql.

```
# filter() # where de sql----------------------------------------------------------------
# Registros que tengan fecha 31 de diciembre
# Antes
vuelos[vuelos$month == 12 & vuelos$day == 31, ]
```

### 4.4 Función mutate()

Una de las funciones más útiles de dplyr es la función mutate(), la cual permite crear nuevas columnas, o modificar las existentes. 

```
# mutate() ----------------------------------------------------------------
# Crear una nueva columna llamada retraso_neto, que contenga la suma de arr_delay y dep_delay

#antes
vuelos$retraso_neto <- vuelos$arr_delay + vuelos$dep_delay

```

```
head(vuelos)
```

```
#ahora
mutate(vuelos, retraso_neto2 = arr_delay + dep_delay)
```

```
# Recordar que esta nueva columna solo existe en la ejecución y no se ha modificado el dataset original.
head(vuelos)

vuelos <- mutate(vuelos, retraso_neto2 = arr_delay + dep_delay)
head(vuelos)
```

```
names(vuelos)
```

```
# Primero defina una nueva data llamada muestra_fl que contenga la fecha, los retrasos, la distancia, y el tiempo de vuelo
muestra_fl <- select(vuelos, 
                      anio,
                      month,
                      day,
                      arr_delay,
                      dep_delay,
                      distance,
                      air_time
                      )
head(muestra_fl)
# En el dataframe anterior, cree una columna llamada gain que contenga la diferenecia entre arr_delay y dep_delay, y otra llamada speed que mida la velocidad en horas (distancia / tiempo de vuelo * 60)
muestra_fl <- mutate(muestra_fl,
                      gain = arr_delay - dep_delay,
                      speed = distance/air_time * 60,
                      id_tabla = "vl" # podemos crear columnas que repitan un valor dado en todas las filas
                      )
```

```
head(muestra_fl)
```

### Operador %>% 

El operador estrella de dplyr. Funciona como una composición de funciones, y nos permite generar _pipe-lines_, permitiendo  que la generación  del código sea más ordenada, y con ello, su lectura más sencilla.

Este operdaor actua de la sifuiente manera, <br>

    x %>% f(y) , es equivalente a,  f(x,y)



```
# Operador %>% ----------------------------------------------------------------
# Ejemplo sencillo 

# head(vuelos)
head(vuelos,10) # 10 primeros registros de vuelos

# Utilizando %>%
vuelos %>% head(10)

```

```
# 1. Genere el conteo de vuelos por mes y día,
# 2. luego filtre por aquellos registros donde el día es 1,
# 3. ordene de manera descendiente por n,
# 4. finalmente muestra la cabecera del dataset resultante.

## ANTES
head(arrange(filter(count(vuelos,month,day),day == 1), desc(n)))
# Cree una columna que repita el máximo de vuelos por día entre los pasos 1  y 2
head(arrange(filter(mutate(count(vuelos,month,day), max_vl = max(n)),day == 1), desc(n)))
```

```
# Genere el conteo de vuelos por mes y día,
# luego filtre por aquellos registros donde el día es 1,
# ordene de manera descendiente por n,
# finalmente muestra la cabecera del dataset resultante.

## AHORA
vuelos %>%
    count(month, day) %>%
    filter(day == 1) %>%
    arrange(desc(n)) %>%
    head()
# Cree una columna que repita el máximo de vuelos por día entre los pasos 1  y 2
vuelos %>%
    count(month, day) %>%
    mutate(max_vl = max(n)) %>%
    filter(day == 1) %>%
    arrange(desc(n)) %>%
    head()
```

```
# A la onsulta anterior podríamos agregar una nueva columna indicando si el total de vuelos es mayor a 970
vuelos %>%
    count(month, day) %>%
    mutate(max_vl = max(n),
           mayor_970 = n > 970 ) %>%
    filter(day == 1) %>%
    arrange(desc(n)) %>%
    head() # Por defecto retorna los 6 primeros registros.
```

## Función ifelse()

Esta función nos permite generar un control de flujo a lo largo de un vector. Por ejemplo:

```
# Si el arr_delay es mayor a 60, la nueva columna llamada "alto_retraso" debe contenerr la palbra "alto", en caso contrario "bajo".
vuelos %>% 
  mutate(alto_retraso = ifelse(arr_delay > 60, "alto", "bajo")) # ifelse(condicion, valor si ocurre condición, valor si no ocurre condición)
```

```
# En vuelos, Genere una nueva variable llamada day2 que pegue un "0" 
# a la variable day si es que esta contiene un dígito del 1 al 9.
# Para pegar cadena de caracteres sin espacios puede utilizar 
# la función paste0

vuelos <- vuelos %>%
  mutate(day2 = ifelse(day <= 9, paste0("0",day), day ),
         month2 = ifelse(month <= 9, paste0("0",month), month ))

#?paste0, ?nchar
paste("2013","01","30", sep = "-")

```

### 4.5 Función group_by() y summarise()

La función `summarise()` calcula medidas agregadas sobre columnas, mientras que la función `group_by`, genera un data.frame "en bloques" dados por la(s) columna(s) ingresadas como parámetros en la función.



```
# group_by() y summarise() ----------------------------------------------------------------
# Ejemplo sencillo
exam2 <- data.frame(
  x = c(1:5),
  y = c(7:11),
  z = c("A", "B", "A", "B", "C")
)
exam2
```

```
data_agrupada <- exam2 %>%
  group_by(z)

data_agrupada
```

```
data_agrupada %>% 
  ungroup()
```

```
# Podemos generar un número de fila por cada grupo
exam2 %>%
  group_by(z) %>%
  mutate( n_fila = row_number())
```

```
exam2 %>%
  mutate(n_fila = row_number())
```

```
exam2 %>%
  group_by(z) %>% # Comparen omitiendo el "group_by"
  mutate( tot_x = sum(x),
          tot_y = sum(y))
```

```
# 1) Aplicar la función summarise a la columna 'y' y calcular su promedio
exam2 %>%
  group_by(z) %>% # Comparen omitiendo el "group_by"
  summarise(tot_x = sum(x),
            tot_y = sum(y),
            prom_x = mean(x))


# 2) Generar grupos según la variable z. (a diferencia de sql, el comando group_by() no requiere de una medida agregada para que este funcione.)
# select z, avg(x) mean_x, sum(x) tot_x from exam2 group by z

```

```
# 3) Calcular media de x e y en cada valor de z.
# podemos calcular máximos y mínimos
exam2 %>%
  group_by(z) %>% # Comparen omitiendo el "group_by"
  summarise(tot_x = sum(x),
            tot_y = sum(y),
            prom_x = mean(x),
            max_x = max(x),
            min_y = min(y))
```

### 4.6 Función sample_n() y sample_frac()
